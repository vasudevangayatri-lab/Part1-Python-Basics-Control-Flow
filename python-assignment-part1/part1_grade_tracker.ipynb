### Task 1 — Data Parsing & Profile Cleaning

```python
# Raw student data provided
raw_students = [
    {"name": "  ayesha SHARMA ", "roll": "101", "marks_str":"88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll":"102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair ",    "roll":"103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai",    "roll":"105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:
   
    clean_name = student["name"].strip().title()
    clean_roll = int(student["roll"])
    clean_marks = [int(m.strip()) for m in student["marks_str"].split(",")]

    cleaned_data = {
        "name": clean_name,
        "roll": clean_roll,
        "marks": clean_marks
    }
    cleaned_students.append(cleaned_data)
  
    is_valid = all(part.isalpha() for part in clean_name.split())
    validation_msg = "- Valid name" if is_valid else "- Invalid name"

    print("=" * 32)
    print(f"Student : {clean_name}")
    print(f"Roll No : {clean_roll}")
    print(f"\nMarks   : {clean_marks}")
    print(f"Status  : {validation_msg}")
    print("=" * 32 + "\n")

# Roll Number 103 processing

print("--- Special Lookup: Roll 103 ---")
for student in cleaned_students:
    if student["roll"] == 103:
        name_upper = student["name"].upper()
        name_lower = student["name"].lower()
        print(f"Uppercase: {name_upper}")
        print(f"Lowercase: {name_lower}")
```

### Task 2 — Marks Analysis Using Loops & Conditionals

```python
# Section 1 — Subject-wise Marks & Grades 

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

def get_grade(score):
    if 90 <= score <= 100:
        return "A+"
    elif 80 <= score <= 89:
        return "A"
    elif 70 <= score <= 79:
        return "B"
    elif 60 <= score <= 69:
        return "C"
    else:
        return "F"

print(f"\n{'='*45}")
print(f"  Grade Report for: {student_name}")
print(f"{'='*45}")
print(f"{'Subject':<12}  {'Marks':>6}  {'Grade':>5}")
print(f"{'-'*45}")

for subject, score in zip(subjects, marks):
    grade = get_grade(score)
    print(f"{subject:<12}  {score:>6}  {grade:>5}")

print(f"{'='*45}")
```

```python
# Section 2 — Calc & Print

total   = sum(marks)
average = round(total / len(marks), 2)

highest_idx = marks.index(max(marks))
lowest_idx  = marks.index(min(marks))

print(f"Statistics for {student_name}")
print(f"{'─'*35}")
print(f"  Total Marks    : {total} / {len(marks) * 100}")
print(f"  Average Marks  : {average}")
print(f"  Highest Score  : {subjects[highest_idx]} — {marks[highest_idx]}")
print(f"  Lowest Score   : {subjects[lowest_idx]}  — {marks[lowest_idx]}")
print(f"{'─'*35}")
```

> Type a subject name and press Enter, then type marks (0–100) and press Enter.
> Type `done` as subject name to stop. Example
>  1. Zoology  - 20
>  2. Cyber security - 85
>  3. Project Management - A10 (Entry not to be counted)
>  4. done  to Exit

```python
# Section 3 — Interactive Marks-Entry System using 'while loop'

print("New Subject Entry System")
print("   (type 'done' as subject name to stop)\n")

new_subjects_added = 0

while True:
    subject_input = input("  Enter subject name : ").strip()

    # Exit condition
    if subject_input.lower() == "done":
        break

    # Empty name guard
    if not subject_input:
        print("  Subject name cannot be empty. Try again.\n")
        continue

    # Marks input
    marks_input = input(f"  Enter marks for {subject_input} (0–100) : ").strip()

    # Check 1: must be numeric
    if not marks_input.replace(".", "", 1).lstrip("-").isdigit():
        print(f"  X Invalid input '{marks_input}' — marks must be a number. Entry skipped.\n")
        continue

    score = float(marks_input)

    # Check 2: must be a whole number
    if score != int(score):
        print(f"  X Marks must be a whole number (got {marks_input}). Entry skipped.\n")
        continue

    score = int(score)

    # Check 3: must be in range 0–100
    if not (0 <= score <= 100):
        print(f"  X Marks {score} is out of range (0–100). Entry skipped.\n")
        continue

    # Valid entry
    subjects.append(subject_input)
    marks.append(score)
    new_subjects_added += 1
    print(f"    {subject_input} ({score}) added successfully.\n")

# Post-loop summary
updated_average = round(sum(marks) / len(marks), 2)

print(f"\n{'='*45}")
print(f"  Entry Session Complete")
print(f"{'='*45}")
print(f"  New subjects added    : {new_subjects_added}")
print(f"  Total subjects now    : {len(subjects)}")
print(f"  Updated average marks : {updated_average}")
print(f"{'='*45}")
```

### Task 3 — Class Performance Summary

```python
# Given student Data

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

```python
# Step 1 — Compute Average and Pass/Fail Status

student_results = []   # stores (name, average, status) for all students

for name, marks in class_data:
    average = round(sum(marks) / len(marks), 2)
    status  = "Pass" if average >= 60 else "Fail"
    student_results.append((name, average, status))
```

```python
# Step 2 — Formatted Class Report

print(f"\n{'Name':<18}| {'Average':^9}| Status")
print("-" * 40)

for name, average, status in student_results:
    print(f"{name:<18}| {average:^9.2f}| {status}")
```

```python
# Step 3 — Pass/Fail Summary 

passed    = [s for s in student_results if s[2] == "Pass"]
failed    = [s for s in student_results if s[2] == "Fail"]
topper    = max(student_results, key=lambda s: s[1])
class_avg = round(sum(s[1] for s in student_results) / len(student_results), 2)

print(f"\n{'─'*40}")
print(f"  Students Passed  : {len(passed)}")
print(f"  Students Failed  : {len(failed)}")
print(f"  Class Topper     : {topper[0]} ({topper[1]})")
print(f"  Class Average    : {class_avg}")
print(f"{'─'*40}")
```

### Task 4 — String Manipulation Utility

```python
# Raw Essay, the given input
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "
```

```python
# Step 1 — Strip Whitespace

clean_essay = essay.strip()
print("Step 1 — Stripped Essay:")
print(clean_essay)
```

```python
# Step 2 — Convert to Title Case

print("Step 2 — Title Case:")
print(clean_essay.title())
```

```python
# Step 3 — Count Occurrences of `"python"`

python_count = clean_essay.count("python")
print("Step 3 — Count of 'python':")
print(f'"python" appears {python_count} times in the essay')
```

```python
# Step 4 — Replace `"python"` with `"Python 🐍"`

print("Step 4 — After Replacing 'python' with 'Python 🐍':")
print(clean_essay.replace("python", "Python 🐍"))
```

```python
# Step 5 — Split the Sentences

sentences = clean_essay.split(". ")
print("Step 5 — Sentences List:")
print(sentences)
```

```python
# Step 6 — Print Numbered Sentences with `.` at the end

print("Step 6 — Numbered Sentences:")
for i, sentence in enumerate(sentences, start=1):
    if not sentence.endswith("."):
        sentence = sentence + "."
    print(f"{i}. {sentence}")
```
